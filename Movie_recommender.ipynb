```python
import requests
import json
import pandas as pd

response=requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1")
json = response.json()
print(json.keys())
# json
```

```python
df=pd.DataFrame((json['results']))
df=df.drop(['adult','backdrop_path','id','original_language','popularity','poster_path','release_date','title','video','vote_average','vote_count'],axis=1)
df.head()
```

```python
for i in range(2,463):
  response=requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}')
  json = response.json()
  temp_df=pd.DataFrame(json['results'])
  temp_df=temp_df.drop(['adult','backdrop_path','id','original_language','popularity','poster_path','release_date','title','video','vote_average','vote_count'],axis=1)
  df = df.append(temp_df, ignore_index=True) # pandas will ignore the existing index labels of the DataFrame and create a new integer index for the combined DataFrame
```

```python
df.head() #we got data of 9233 movies sucessfully
```

```python
#Now we have to change genre_ids to orginal names
genres=requests.get('https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US')
x=genres.json()['genres']
x
```

```python
# Create a mapping dictionary from genre IDs to genre names
id_to_name = {genre['id']: genre['name'] for genre in x}

# Convert genre_ids to genre names
df['genres'] = df['genre_ids'].apply(lambda ids: [id_to_name[id] for id in ids])
df.head()
```

```python
df.drop('genre_ids',axis=1,inplace=True)
df.head()
```
